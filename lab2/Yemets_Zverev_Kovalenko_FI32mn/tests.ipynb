```python
import random

def inbuild_generator(n):
  massive = [random.randint(0,1) for x in range(n)]
  return massive

def L20_generator(n):
  massive = [random.randint(0,1) for x in range(20)]
  for i in range(n-20):
    massive.append((massive[-3]+massive[-5]+massive[-9]+massive[-20])%2)
  return massive

def L89_generator(n):
  massive = [random.randint(0,1) for x in range(89)]
  for i in range(n-89):
    massive.append((massive[-38] + massive[-89])%2)
  return massive


def LehmerLow(x0):
  a = 2**16 + 1
  m = 2**32
  c = 119
  x_next = (a*x0 + c)%m
  x_next = bin(x_next)
  x_next = list(x_next)
  for i in range(2):
    x_next.pop(0)
  while len(x_next) < 32:
    x_next.insert(0,0)
  result = x_next[-8:]
  result = [int(elem) for elem in result]
  return result

def LehmerHigh(x0):
  a = 2**16 + 1
  m = 2**32
  c = 119
  x_next = (a*x0 + c)%m
  x_next = bin(x_next)
  x_next = list(x_next)
  for i in range(2):
    x_next.pop(0)
  while len(x_next) < 32:
    x_next.insert(0,0)
  result = x_next[0:8]
  result = [int(elem) for elem in result]
  return result

def L11(vector):
  copy = vector.copy()
  next = copy[0]^copy[2]
  delete = copy[0]
  copy.pop(0)
  copy.append(next)
  return delete, copy

def L9(vector):
  copy = vector.copy()
  next = copy[0]^copy[1]^copy[3]^copy[4]
  delete = copy[0]
  copy.pop(0)
  copy.append(next)
  return delete, copy

def L10(vector):
  copy = vector.copy()
  next = copy[0]^copy[3]
  delete = copy[0]
  copy.pop(0)
  copy.append(next)
  return delete, copy

def function(vector_11, vector_9, vector_10):
  x_i = vector_11
  y_i = vector_9
  s_i = vector_10
  z_i = (s_i*x_i + y_i*(1+s_i)%2)%2
  return z_i

def Geffe(vector1, vector2, vector3):
  x_i, vector11_n = L11(vector1)
  y_i, vector9_n = L9(vector2)
  s_i, vector10_n = L10(vector3)
  z_i = function(x_i, y_i, s_i)
  return z_i, vector11_n, vector9_n, vector10_n

def Bluma_Mikali(n):
  p = 0xCEA42B987C44FA642D80AD9F51F10457690DEF10C83D0BC1BCEE12FC3B6093E3
  a = 0x5B88C41246790891C095E2878880342E88C79974303BD0400B090FE38A688356
  q = 0x675215CC3E227D3216C056CFA8F8822BB486F788641E85E0DE77097E1DB049F1
  from random import randint
  T = randint(0, p - 1)
  result = []
  if T < q:
    result.append(1)
  else:
    result.append(0)
  for i in range(n-1):
    T = pow(a, T, p)
    if T < q:
      result.append(1)
    else:
      result.append(0)
  return result

def Bluma_Mikali_Bytes(n):
  p = 0xCEA42B987C44FA642D80AD9F51F10457690DEF10C83D0BC1BCEE12FC3B6093E3
  a = 0x5B88C41246790891C095E2878880342E88C79974303BD0400B090FE38A688356
  q = 0x675215CC3E227D3216C056CFA8F8822BB486F788641E85E0DE77097E1DB049F1
  from random import randint
  T = randint(0, p - 1)
  result = []
  for i in range(n):
    T = pow(a, T, p)
    for res in range(256):
      if (res * (p-1) < 256*T and 256*T <= (res+1)*(p-1)):
        result.append(res)
      else:
        continue
  return result

def BBS(r0, length):
  p = 0xD5BBB96D30086EC484EBA3D7F9CAEB07
  q = 0x425D2B9BFDB25B9CF6C416CC6E37B59C1F
  n = p*q
  result = []
  r = (r0**2)%n
  x_new = r%2
  result.append(x_new)
  for i in range(length-1):
    r = (r**2)%n
    x = r%2
    result.append(x)
  return result

def BBS_Bytes(r0, length):
  p = 0xD5BBB96D30086EC484EBA3D7F9CAEB07
  q = 0x425D2B9BFDB25B9CF6C416CC6E37B59C1F
  n = p*q
  result = []
  r = (r0**2)%n
  x_new = r%256
  result.append(x_new)
  for i in range(length-1):
    r = (r**2)%n
    x = r%256
    result.append(x)
  return result

def Wolfram(r0):
  temp = r0.copy()
  temp.append(temp[0])
  temp.pop() # Циклический сдвиг влево
  r0_left = temp.copy() # Уже сдвинутый влево массив
  temp = r0.copy()
  temp.insert(temp[-1], 0)
  temp.pop(-1)
  r0_right = temp.copy() # Уже сдвинутый вправо массив
  r0_massive = r0.copy()
  result = []
  for i in range(32):
    result.append(r0_left[i]^(r0_massive[i] | r0_right[i]))
  return result

def Librarian(text, n):
    file = open(text, encoding = 'utf-8')
    massive = []
    for row in file:
      for symbol in row:
        if len(massive)<n:
          massive.append(ord(symbol))
        else:
          break
      if len(massive)>=n:
        break
    return massive
```
