```python
import time
import timeit

A = int('5B88C41246790891C095E2878880342E88C79974303BD0400B090FE38A688356', 16)
B = int('675215CC3E227D3216C056CFA8F8822BB486F788641E85E0DE77097E1DB049F1', 16)
P = int('CEA42B987C44FA642D80AD9F51F10457690DEF10C83D0BC1BCEE12FC3B6093E3', 16)
parametres = {'A': A, 'B': B, 'P': P}
n = 10000

C = A + B
print(hex(C)[2:])
cod = """
C = A + B
"""
start1 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish1 = timeit.default_timer()
print(f"Час виконання операції +: {(finish1 - start1) / n * 1e6} мікросекунд")

D = B - A
print(hex(D)[2:])
cod = """
D = B - A
"""
start2 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish2 = timeit.default_timer()
print(f"Час виконання операції -: {(finish2 - start2) / n * 1e6} мікросекунд")

E = A ** 2
print(hex(E)[2:])
cod = """
E = A ** 2
"""
start3 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish3 = timeit.default_timer()
print(f"Час виконання операції ^2: {(finish3 - start3) / n * 1e6} мікросекунд")

F = A * B
print(hex(F)[2:])
cod = """
F = A * B
"""
start4 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish4 = timeit.default_timer()
print(f"Час виконання операції *: {(finish4 - start4) / n * 1e6} мікросекунд")

J = F // A
print(hex(J)[2:])
cod = """
J = F // A
"""
start5 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish5 = timeit.default_timer()
print(f"Час виконання операції /: {(finish5 - start5) / n * 1e6} мікросекунд")

H = A % P
print(hex(H)[2:])
cod = """
H = A % P
"""
start6 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish6 = timeit.default_timer()
print(f"Час виконання операції mod: {(finish6 - start6) / n * 1e6} мікросекунд")

L = pow(A, B, P)
print(hex(L)[2:])
cod = """
L = pow(A, B, P)
"""
start7 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish7 = timeit.default_timer()
print(f"Час виконання операції A ^ B mod P: {(finish7 - start7) / n * 1e6} мікросекунд")

M = pow(A, -1, P)
print(hex(M)[2:])
cod = """
M = pow(A, -1, P)
"""
start8 = timeit.default_timer()
for _ in range(n):
    exec(cod, parametres)
finish8 = timeit.default_timer()
print(f"Час виконання операції A ^ -1 mod P: {(finish8 - start8) / n * 1e6} мікросекунд")
```
